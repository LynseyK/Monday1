```python
# -----------------------
# 1) Define each room as a dictionary of directions
# -----------------------
Cafeteria = {
    "north": "Nothing",
    "east":  "Nothing",
    "south": "Nothing",
    "west":  "Hallway_south"
}

Admin = {
    "north": "Nothing",
    "east":  "Navigation",
    "south": "Nothing",
    "west":  "Hallway_middle"
}

Electrical = {
    "north": "Nothing",
    "east":  "Storage",
    "south": "Nothing",
    "west":  "Hallway_north"
}

Hallway_south = {
    "north": "Hallway_middle",
    "east":  "Cafeteria",
    "south": "Nothing",
    "west":  "Nothing"
}

Hallway_middle = {
    "north": "Hallway_north",
    "east":  "Admin",
    "south": "Hallway_south",
    "west":  "Nothing"
}

Hallway_north = {
    "north": "Nothing",
    "east":  "Electrical",
    "south": "Hallway_middle",
    "west":  "Medbay"
}

Medbay = {
    "north": "Nothing",
    "east":  "Hallway_north",
    "south": "Nothing",
    "west":  "Nothing"
}

Navigation = {
    "north": "Storage",
    "east":  "Nothing",
    "south": "Nothing",
    "west":  "Admin"
}

Storage = {
    "north": "Nothing",
    "east":  "Nothing",
    "south": "Navigation",
    "west":  "Admin"
}

# -----------------------
# 2) Combine these into a single map/dict for easy lookup
# -----------------------
room_map = {
    "Cafeteria": Cafeteria,
    "Admin": Admin,
    "Electrical": Electrical,
    "Hallway_south": Hallway_south,
    "Hallway_middle": Hallway_middle,
    "Hallway_north": Hallway_north,
    "Medbay": Medbay,
    "Navigation": Navigation,
    "Storage": Storage
}

# -----------------------
# 3) Define the move function
# -----------------------
def move_player_dictionary(current_room: str, direction: str) -> str:
    """
    Moves the player to a new room based on dictionary lookups of directions.
    
    :param current_room: Name of the current room (e.g., "Hallway_south")
    :param direction: One of "north", "east", "south", "west"
    :return: The name of the new room if valid, or the same room if it's "Nothing"
    """
    # Get the dictionary of exits for the current room
    exits = room_map[current_room]
    
    # Get the new room based on the direction
    new_room = exits.get(direction, "Nothing")
    
    # If the room is valid (not "Nothing"), move there; otherwise stay put
    if new_room != "Nothing":
        return new_room
    else:
        return current_room

# -----------------------
# 4) Main loop to prompt the user for input
# -----------------------
def main():
    # Starting position
    player_position = "Hallway_south"
    print("You are in:", player_position)
    
    while True:
        # Prompt user for direction
        user_input = input("\nWhich direction do you want to go? (north/east/south/west or quit): ").strip().lower()
        
        # Check if the user wants to quit
        if user_input == "quit":
            print("Goodbye!")
            break
        
        # Validate the input
        if user_input not in ["north", "east", "south", "west"]:
            print("Invalid direction. Please enter north, east, south, west, or quit.")
            continue
        
        # Attempt to move
        old_position = player_position
        player_position = move_player_dictionary(player_position, user_input)
        
        # Check if movement happened
        if player_position == old_position:
            print("You can't go that way!")
        else:
            print("You move", user_input, "to:", player_position)

# -----------------------
# 5) Run if executed as a script
# -----------------------
if __name__ == "__main__":
    main()
```
